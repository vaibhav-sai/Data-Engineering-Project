### GOLD LAYER NOTEBOOK: TRANSFORMANTIONS AND AGGREGATIONS SHOPPINGMART DATA

```python
from pyspark.sql.functions import *
```

### Read Data from silver layer in parquet files into dataframes

```python
Orders_df = spark.read.parquet("abfss://292fd850-c68a-4bad-b4f3-7cabce198e06@onelake.dfs.fabric.microsoft.com/e4be63a0-b193-4799-9364-384d3808141f/Files/ShoppingMart_Silver_Orders/ShoppingMart_customers_orderdata")
reviews_df = spark.read.parquet("abfss://292fd850-c68a-4bad-b4f3-7cabce198e06@onelake.dfs.fabric.microsoft.com/e4be63a0-b193-4799-9364-384d3808141f/Files/ShoppingMart_Silver_Reviews/ShoppingMart_review")
social_df = spark.read.parquet("abfss://292fd850-c68a-4bad-b4f3-7cabce198e06@onelake.dfs.fabric.microsoft.com/e4be63a0-b193-4799-9364-384d3808141f/Files/ShoppingMart_Silver_Social_Media/ShoppingMart_social_media")
weblogs_df = spark.read.parquet("abfss://292fd850-c68a-4bad-b4f3-7cabce198e06@onelake.dfs.fabric.microsoft.com/e4be63a0-b193-4799-9364-384d3808141f/Files/ShoppingMart_Silver_Web_Logs/ShoppingMart_web_logs")
```

### Aggregates web log data to measure engagement per user on each page and action.

```python
weblogs_df = weblogs_df.groupBy("user_id", "page", "action").count()
display(weblogs_df)
```

```python
weblogs_df.write.format("delta").mode("overwrite").saveAsTable("Shopping_Mart_Gold_Schema.ShoppingMart_web_logs")
```

### Aggregates unstructured social media data to track sentiment trends across different platforms

```python
social_df= social_df.groupBy("platform","sentiment" ).count()
display(social_df)
```

```python
social_df.write.format("delta").mode("overwrite").saveAsTable("Shopping_Mart_Gold_Schema.ShoppingMart_social_media")
```

### Aggregates product reviews to calculate the average rating per product.

```python
reviews_df = reviews_df.groupBy("product_id").agg(avg("rating").alias("AvgRating"))
display(reviews_df)
```

```python
reviews_df.write.format("delta").mode("overwrite").saveAsTable("Shopping_Mart_Gold_Schema.ShoppingMart_review")
```
